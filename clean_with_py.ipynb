```python
df.columns = df.columns.str.lower().str.strip()
dd = df
dd
```

```python
dd["orderdate"] = pd.to_datetime(dd["orderdate"],format ="mixed")
```

```python
dd["orderdate"]
```

```python
dd["unitprice"] = (
    dd["unitprice"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", ".", regex=False)
)

dd["unitprice"] = pd.to_numeric(dd["unitprice"], errors="coerce")
```

```python
dd["unitprice"]
```

```python
dd["region"] = dd["region"].fillna("unknown")
```

```python
dd["channel"] = dd["channel"].fillna("unknown")
```

```python
dd["customer"] = dd["customer"].fillna("unknown")
```

```python
dd["quantity"] = dd["quantity"].fillna(
    dd.groupby("product")["quantity"].transform("median")
)
```

```python
cols = ["product","region","channel","customer"]
```

```python
for col in cols:
    dd[col] = dd[col].str.strip().str.title()
```

```python
dd
```

```python
dd.loc[df["quantity"]<0,"quantity"]=dd.loc[df["quantity"]<0,"quantity"].abs()
```
